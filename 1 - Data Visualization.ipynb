```python
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread(r'Malaria Cell Images Dataset\Uninfected\C1_thinF_IMG_20150604_104722_cell_9.png')
image = cv2.resize(image, (180,180), interpolation=cv2.INTER_CUBIC)
plt.imshow(image)
plt.title('Uninfected')
plt.axis('off')
plt.show()

image = cv2.imread(r'Malaria Cell Images Dataset\Parasitized\C33P1thinF_IMG_20150619_114756a_cell_179.png')
image = cv2.resize(image, (180,180), interpolation=cv2.INTER_CUBIC)
plt.imshow(image)
plt.title('Parasitized')
plt.axis('off')
plt.show()
```

```python
negativeVals = len(os.listdir(r'C:\Users\raami\Downloads\Research Paper Works\Contract-3\Malaria Cell Images Dataset\Uninfected'))
positiveVals = len(os.listdir(r'C:\Users\raami\Downloads\Research Paper Works\Contract-3\Malaria Cell Images Dataset\Parasitized'))


plt.bar(['Positive', 'Negative'], [positiveVals, negativeVals], color=['r', 'g'])
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Image Count for each class')
plt.show()


X = ['Training','Validation']
posCount = [positiveVals*0.8, positiveVals*0.2]
negCount = [negativeVals*0.8, negativeVals*0.2]
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2, posCount, 0.4, label = 'Positive Class', color='r')
plt.bar(X_axis + 0.2, negCount, 0.4, label = 'Negative Class', color='g')
plt.xticks(X_axis, X)
plt.xlabel("Category of Data")
plt.ylabel("Count")
plt.title("Splitting of Data into Training and Testing")
plt.legend()
plt.show()
```
